```python
from ipynb.fs.full.Preprocessing import ReturnData,getData
from sklearn import metrics
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
import time
```

```python
allData = ReturnData()

TrainX = np.concatenate((allData[0][0],allData[1][0]),axis=0)
TrainY = np.concatenate((allData[0][1],allData[1][1]),axis=0)
TestX = allData[2][0]
TestY = allData[2][1]
```

## Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
s = time.time()
DT = DecisionTreeClassifier()

DT.fit(TrainX,TrainY.astype('int'))

TrainingPrediction = DT.predict(TrainX)
e = time.time()
print("The Training Confusion Matrix is: \n", metrics.confusion_matrix(TrainingPrediction,TrainY.astype('int')),"\n")
print("The Training Accuracy is: ", metrics.accuracy_score(TrainingPrediction,TrainY.astype('int'))*100,"%\n")
print("The Time taken is:",e-s)
```

```python
TestPrediction = DT.predict(TestX)

print("The Test Confusion Matrix is: \n", metrics.confusion_matrix(TestPrediction,TestY.astype('int')),"\n")
print("The Test Accuracy is: ", metrics.accuracy_score(TestPrediction,TestY.astype('int'))*100,"%\n")
print("The Test Precision is: ", metrics.precision_score(TestPrediction,TestY.astype('int'),average=None)*100,"%\n")
print("The Test Recall is: ", metrics.recall_score(TestPrediction,TestY.astype('int'),average=None)*100,"%\n")
```

```python
plt.plot(np.arange(len(TestY)),TestY,label = "Actual")
plt.plot(np.arange(len(TestY)),TestPrediction,label = "Predicted",color="red")
plt.xlabel("data-point")
plt.ylabel("target")
plt.title("Predicted VS Actual")
plt.legend(loc = 'upper right')
plt.show()
plt.close()
```

## Support Vector Machines

```python
from sklearn.svm import SVC
```

```python
start = time.time()
svm = SVC(kernel='linear')
svm.fit(TrainX,TrainY.astype('int'))

TrainingPrediction = svm.predict(TrainX)
end = time.time()
print("The Training Confusion Matrix is: \n", metrics.confusion_matrix(TrainingPrediction,TrainY.astype('int')),"\n")
print("The Training Accuracy is: ", metrics.accuracy_score(TrainingPrediction,TrainY.astype('int'))*100,"%\n")
print("The Time taken is:",end-start)
```

```python
TestPrediction = svm.predict(TestX)

print("The Test Confusion Matrix is: \n", metrics.confusion_matrix(TestPrediction,TestY.astype('int')),"\n")
print("The Test Accuracy is: ", metrics.accuracy_score(TestPrediction,TestY.astype('int'))*100,"%\n")
print("The Test Precision is: ", metrics.precision_score(TestPrediction,TestY.astype('int'),average=None)*100,"%\n")
print("The Test Recall is: ", metrics.recall_score(TestPrediction,TestY.astype('int'),average=None)*100,"%\n")
```

```python
plt.plot(np.arange(len(TestY)),TestY,label = "Actual")
plt.plot(np.arange(len(TestY)),TestPrediction,label = "Predicted",color="red")
plt.xlabel("data-point")
plt.ylabel("target")
plt.title("Predicted VS Actual")
plt.legend(loc = 'upper right')
plt.show()
plt.close()
```
